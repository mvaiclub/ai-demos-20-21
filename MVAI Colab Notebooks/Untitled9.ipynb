```
#what does each import do?
import pandas as pd
import tensorflow as tf

#what do the next two lines do?
train = pd.read_csv("sign_mnist_train.csv")
test = pd.read_csv("sign_mnist_test.csv")

#what do the next two lines do?
ytrain = train['label'].values
ytest = test['label'].values

# what does "drop" mean and what is it doing?
train.drop('label', axis = 1, inplace = True)
test.drop('label', axis = 1, inplace = True)

#rename vars
xtrain = train
xtest = test

#what does to_categorical do and why do we need it?
ytrain = tf.keras.utils.to_categorical(ytrain)
ytest = tf.keras.utils.to_categorical(ytest)

#what are these lines for?
xtrain = xtrain/255.0
xtest = xtest/255.0

#what does reshape do and why have we chosen these numbers?
xtrain = xtrain.values.reshape(len(ytrain), 28,28,1)
xtest = xtest.values.reshape(len(ytest), 28,28,1)

k = tf.keras
model = k.Sequential()

#what does Conv2D mean/what does it stand for? what does (3,3) mean? 
model.add(k.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))

#challenge: what does MaxPooling2D do?
model.add(k.layers.MaxPooling2D((2, 2)))

#what does flatten do?
model.add(k.layers.Flatten())

#what does Dense mean here? what deos the 100 mean?
model.add(k.layers.Dense(100, activation='relu'))

#what is this layer for?
model.add(k.layers.Dense(25, activation='softmax'))

#what does loss mean?
model.compile(optimizer="Adam", loss='categorical_crossentropy', metrics=['accuracy'])

model.fit(xtrain, ytrain, epochs=5)
print(model.evaluate(xtest, ytest)[0])
```
