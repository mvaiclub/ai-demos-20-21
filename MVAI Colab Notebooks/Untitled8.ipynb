```
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```
#get data
data = pd.read_csv("BankChurners.csv")

data.head()

```

```
#remove unnecessary  two columns
data.drop(columns=data.columns[-2:], inplace=True)
data.drop(columns=data.columns[0],axis=1, inplace=True)
data.head()
```

```
# Finding all the categorical columns from the data
categorical = data.select_dtypes(exclude=['int64','float64']).columns
for i in categorical:
    data[i]=pd.factorize(data[i])[0]
data.head()

#normalize data
for col in data:
  data[col] = data[col]/data[col].max()
```

```
#divide input and output
y=data['Attrition_Flag'].values
x=data.drop('Attrition_Flag',axis=1).values

#train test split
xtrain, xtest, ytrain, ytest = train_test_split( x, y, test_size=0.2, random_state=20,shuffle=True)
```

```
k = tf.keras.layers
model = tf.keras.Sequential()

model.add(k.Dense(100, activation = 'relu'))
model.add(k.Dense(1, activation = 'sigmoid'))

model.compile(optimizer = "adam", loss = "binary_crossentropy", metrics = ['accuracy'])
model.fit(xtrain, ytrain, epochs = 50)

print(model.evaluate(xtest, ytest)[1])
```
