```
import pandas as pd
import tensorflow as tf
import numpy as np
```

```
data = pd.read_csv("Reviews.csv", engine= "python", error_bad_lines=False)
data = data.head(15000)

print(data.shape)
```

```
data.dropna(how='any', inplace=True)
data.reset_index(drop=True)

data =  pd.DataFrame(data, columns=['Score', 'Text'])




data.head()
print(data["Text"].head())

import string
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
summary_data = data["Text"].values
#text_data.pop(25509)
#print(text_data[2550])
for i in range(len(summary_data)):
  summary_data[i] = summary_data[i].lower()
  summary_data[i] = "".join([char for char in summary_data[i] if char not in string.punctuation])
  summary_data[i] = " ".join([word for word in str(summary_data[i]).split() if word not in stop_words])

```

```
print(data["Text"][16])
```

```
print(summary_data)
```

```

print(data["Score"][2])
score_data = data["Score"].values
score_data = score_data.astype("float32")
print(score_data)
for i in range(len(score_data)):
  score_data[i] -= 1.0
print(score_data[1])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(summary_data, score_data, test_size = 0.4, random_state = 34)

import scipy.sparse

from scipy.sparse import csr_matrix
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
#data['Text'] = count_vect.fit(data['Text'].values)
print("hi")
xtrain = count_vect.fit_transform(xtrain)
xtest = count_vect.transform(xtest)
print(count_vect.vocabulary_)

xtrain = xtrain.toarray()
xtest = xtest.toarray()

# for x in summary_data:
#   print(summary_data[x])
#   summary_data[x] = csr_matrix.toarray(summary_data[x])
print(xtrain)
```

```


k = tf.keras
model = k.Sequential()
#print(score_data.shape)

#xtrain = np.expand_dims(xtrain, 1)
#xtest = np.expand_dims(xtest, 1)
#print(xtrain.shape)

#what does Conv2D mean/what does it stand for? what does (3,3) mean? 
#model.add(k.layers.LSTM(50, input_shape = (xtrain.shape[1:]), return_sequences = False))

#what does Dense mean here? what deos the 100 mean?
model.add(k.layers.Dense(2048, activation='relu'))
model.add(k.layers.Dropout(0.5))
model.add(k.layers.Dense(1024, activation='relu'))
model.add(k.layers.Dropout(0.5))
model.add(k.layers.Dense(512, activation='relu'))
model.add(k.layers.Dropout(0.5))
#what is this layer for?
model.add(k.layers.Dense(5, activation='softmax'))

#what does loss mean?
model.compile(optimizer="Adam", loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])

model.fit(xtrain, ytrain, epochs=20, validation_data= (xtest, ytest) )
```

```
model.evaluate(xtest, ytest)
```
